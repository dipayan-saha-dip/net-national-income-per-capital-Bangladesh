```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/gdrive/MyDrive/Dataset/CodeBasis Project/Machine Learning Tutorial/1/National income per capita Bangladesh.csv')
df.head(15)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(df.year,df.income)
```

```python
print(df.shape)
df.dtypes
```

```python
df.describe()
```

```python
new_df = df.drop('income',axis='columns')
new_df
```

```python
income = df.income
income
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,income)
```

```python
reg.predict([[2023]])
```

```python
reg.predict([[2030]])
```

```python
years_future = np.arange(2021, 2031)
income_pred = reg.predict(years_future[:, np.newaxis])
```

```python
plt.scatter(df.year, df.income)
plt.plot(years_future, income_pred, color='red')
plt.title('Bangladesh National Income per Capita')
plt.xlabel('Year')
plt.ylabel('Income ($)')
plt.show()
```

```python
print("Predicted Income:")
for i, pred in enumerate(income_pred):
    print(f"{2021 + i}: {pred:.2f}")
```
